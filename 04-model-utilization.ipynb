#### Import Libraries

```python
import numpy as np
import pandas as pd
import warnings
import pickle as pkl
warnings.filterwarnings('ignore')
```

#### User Input

```python
age = int(input("Enter Age: "))
sex = input("Enter Sex (male/female): ")
bmi = float(input("Enter BMI: "))
children = int(input("Enter Number of Children: "))
smoker = input("Enter Smoker (yes/no): ")
region = input("Enter Region: ")
```

#### Create DataFrame

```python
data = [[age, sex, bmi, children, smoker, region]]

columns = ['age', 'sex', 'bmi', 'children', 'smoker', 'region']

myinput = pd.DataFrame(data=data, columns=columns)

myinput
```

#### Load Model & Predict

```python
pipe = pkl.load(open("hospital_pipe.pkl", "rb"))

result = pipe.predict(myinput)

print("Predicted Hospital Treatment Cost is:", round(result[0]))
```
